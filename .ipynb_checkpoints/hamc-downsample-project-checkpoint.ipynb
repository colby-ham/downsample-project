```python
# function to create the downsampled files
def create_downsample():
    for i in range(1,6):
        print("Creating Downsample for file #" + str(i))
        
        testfile_name = "sgpmetE13.b1.2018010" + str(i) + ".000000.cdf"
        outfile_name = "sgpmetavgE13.b1.2018010" + str(i) + ".000000.cdf"
        
        #remove file if exists
        try:
            os.remove(outfile_name)
        except OSError:
            pass
        
        #open file to downsample
        testfile = Dataset(testfile_name, "r", format="NETCDF4")
        
        #temp vars to help with downsample
        n_atmos_pressure = testfile['atmos_pressure'].shape[0]
        n_temp_mean = testfile['temp_mean'].shape[0]
        n_ap = int(n_atmos_pressure / 5)
        n_tm = int(n_temp_mean / 5)
        cp_ap = np.copy(testfile["atmos_pressure"])
        cp_tm = np.copy(testfile["temp_mean"])
        rs_cp_ap = cp_ap.reshape((n_ap, 5))
        rs_cp_tm = cp_tm.reshape((n_tm, 5))
        
        #downsample
        mean_ap = np.mean(rs_cp_ap, axis=1, dtype=np.float64)
        mean_tm = np.mean(rs_cp_tm, axis=1, dtype=np.float64)
        
        #reconvert to float32
        mean_ap = mean_ap.astype(np.float32)
        mean_tm = mean_tm.astype(np.float32)
        
        #slice time to be correct for downsample
        sliced_time = testfile['time'][::5]
        sliced_time_offset = testfile['time'][::5]
        
        #create new netcdf4 file to output the downsample
        outfile = Dataset(outfile_name, "w", format="NETCDF4")
        outfile.createDimension('time', size = None)
        
        #debug printing 
        #print(outfile.dimensions)
        #print(outfile.dimensions.keys())
        #print(outfile.dimensions['time'])
        
        #copy variable information from old file
        # and replace with the new appropriate data
        temp_obj = testfile['base_time'][0]
        base_time = outfile.createVariable("base_time", temp_obj.dtype, fill_value = False)
        base_time[0] = temp_obj
        base_time.string = testfile['base_time'].string
        base_time.long_name = testfile['base_time'].long_name
        base_time.units = testfile['base_time'].units

        temp_obj = testfile['time_offset'][0]
        time_offset = outfile.createVariable("time_offset", temp_obj.dtype, fill_value=False, dimensions=testfile['time_offset'].dimensions)
        time_offset[:] = sliced_time_offset
        time_offset.long_name = testfile['time_offset'].long_name
        time_offset.units = testfile['time_offset'].units

        temp_obj = testfile['time'][0]
        time = outfile.createVariable("time", temp_obj.dtype, fill_value=False, dimensions=testfile['time'].dimensions)
        time[:] = sliced_time
        time.long_name = testfile['time'].long_name
        time.units = testfile['time'].units
        time.standard_name = testfile['time'].standard_name

        temp_obj = testfile['atmos_pressure'][0]
        atmospheric_pressure = outfile.createVariable("atmospheric_pressure", temp_obj.dtype, fill_value=False, dimensions=testfile['atmos_pressure'].dimensions)
        atmospheric_pressure[:] = mean_ap
        atmospheric_pressure.long_name = testfile['atmos_pressure'].long_name
        atmospheric_pressure.units = testfile['atmos_pressure'].units
        atmospheric_pressure.valid_min = testfile['atmos_pressure'].valid_min
        atmospheric_pressure.valid_max = testfile['atmos_pressure'].valid_max
        atmospheric_pressure.valid_delta = testfile['atmos_pressure'].valid_delta
        atmospheric_pressure.missing_value = testfile['atmos_pressure'].missing_value

        temp_obj = testfile['temp_mean'][0]
        mean_temperature = outfile.createVariable("mean_temperature", temp_obj.dtype, fill_value=False, dimensions=testfile['temp_mean'].dimensions)
        mean_temperature[:] = mean_tm
        mean_temperature.long_name = testfile['temp_mean'].long_name
        mean_temperature.units = testfile['temp_mean'].units
        mean_temperature.valid_min = testfile['temp_mean'].valid_min
        mean_temperature.valid_max = testfile['temp_mean'].valid_max
        mean_temperature.valid_delta = testfile['temp_mean'].valid_delta
        mean_temperature.missing_value = testfile['temp_mean'].missing_value

        temp_obj = testfile['lat'][0]
        lat = outfile.createVariable("lat", temp_obj.dtype, fill_value = False)
        lat[0] = temp_obj
        lat.long_name = testfile['lat'].long_name
        lat.units = testfile['lat'].units
        lat.valid_min = testfile['lat'].valid_min
        lat.valid_max = testfile['lat'].valid_max
        lat.standard_name = testfile['lat'].standard_name

        temp_obj = testfile['lon'][0]
        lat = outfile.createVariable("lon", temp_obj.dtype, fill_value = False)
        lat[0] = temp_obj
        lat.long_name = testfile['lon'].long_name
        lat.units = testfile['lon'].units
        lat.valid_min = testfile['lon'].valid_min
        lat.valid_max = testfile['lon'].valid_max
        lat.standard_name = testfile['lon'].standard_name

        temp_obj = testfile['alt'][0]
        alt = outfile.createVariable("alt", temp_obj.dtype, fill_value = False)
        alt[0] = temp_obj
        alt.long_name = testfile['alt'].long_name
        alt.units = testfile['alt'].units
        alt.standard_name = testfile['alt'].standard_name

        #print(outfile.variables)
        testfile.close()
        outfile.close()
    print("Downsample of all files complete")
    
```

```python
# call function to create the downsampled files
create_downsample()
```

```python
# unit tests
testing_down = Dataset("sgpmetavgE13.b1.20180102.000000.cdf", "r", format="NETCDF4")
testing_original = Dataset("sgpmetE13.b1.20180102.000000.cdf", "r", format="NETCDF4")

# 5 mins apart in downsample, 1 min apart in original file
test_spacing1 = testing_down['time'][1] - testing_down['time'][0]
test_spacing2 = testing_down['time'][2] - testing_down['time'][1]

test_spacing3 = testing_original['time'][1] - testing_original['time'][0]
test_spacing4 = testing_original['time'][2] - testing_original['time'][1]

assert(test_spacing1 == 300.0)
assert(test_spacing2 == 300.0)
assert(test_spacing3 == 60.0)
assert(test_spacing4 == 60.0)
assert((test_spacing1 / test_spacing3) == 5.0)
assert((test_spacing2 / test_spacing4) == 5.0)

# assert that there are 1/5 the entries of the original file
#  and that the vars are the same length
n_down_ap = testing_down['atmospheric_pressure'].shape[0]
n_down_mt = testing_down['mean_temperature'].shape[0]

n_original_ap = testing_original['atmos_pressure'].shape[0]
n_original_mt = testing_original['temp_mean'].shape[0]

assert(n_original_ap / 5 == n_down_ap)
assert(n_original_mt / 5 == n_down_mt)
assert(n_original_ap == n_original_mt)
assert(n_down_ap == n_down_mt)


testing_down.close()
testing_original.close()
print("Tests passed!")
```
